```python
linear_extrap_feat = ['request_abs', 'request_relative', 'interfax_china', 'interfax_tourism', 'interfax_shops',
                     'interfax_medicine', 'google_trend_cov']

#Ð¿Ñ€Ð¸ Ñ…Ð¾Ñ€Ð¾ÑˆÐµÐ¼ ÑÑ†ÐµÐ½Ð°Ñ€Ð¸Ð¸ ÑÑ‚Ð¸ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ ÑÑ‚Ð¸Ñ… Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð¾Ð² Ð±ÑƒÐ´ÑƒÑ‚ ÑÐ½Ð¸Ð¶Ð°Ñ‚ÑŒÑÑ
notlinear_extrap_feat_down = ['isolation_index', 'interfax_isolation']
```

```python
covid_feat = pd.read_csv('good_scenario.csv')
covid_feat = covid_feat[['infected', 'recovered', 'deaths']]
```

```python
for col in covid_feat.columns:
    covid_feat[col] = covid_feat[col].astype(np.int64)
```

```python
numweek_extrap = 12
infected_extrap = covid_feat['infected'][:numweek_extrap].to_numpy()
product_detail_feature = ['price']
notlin_product_detail_feat = ['pctlpt_1_stock', 'item_request']
```

```python
#ÑÐ¾Ð±Ð¸Ñ€Ð°ÐµÐ¼ Ð´Ð°Ñ‚Ð°ÑÐµÑ‚ Ð´Ð»Ñ Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð¾Ð², ÐºÐ¾Ñ‚Ð¾Ñ€Ñ‹Ðµ Ð·Ð°Ð²Ð¸ÑÑÑ‚ Ð¾Ñ‚ Ð½Ð¾Ð¼ÐµÑ€Ð° Ð½ÐµÐ´ÐµÐ»Ð¸
max_growth = 0.2
bias = 0
d_start = df['curr_week'].max()
dd = [d_start + timedelta(days=x) for x in range(7, 7*(numweek_extrap+1), 7)]
test_0 = pd.DataFrame(data = {'curr_week': dd})
for col_name in linear_extrap_feat:
    x = df[(df['product_code'] == 10065) & (df['week_infected'] > 0)]['infected_cumsum']
    y = df[(df['product_code'] == 10065) & (df['week_infected'] > 0)][col_name]
    f = interpolate.interp1d(x, y, fill_value='extrapolate')
    xnew = np.array(infected_extrap)
    ynew = f(xnew)
    
    if ynew[1] <= ynew[0]:
        ynew_min = y.min()
        ynew[ynew < ynew_min] = ynew_min
    else:
        ynew_max = y.max() * (1+max_growth)
        ynew[ynew > ynew_max] = ynew_max
    test_0[col_name] = ynew

for col_name in notlinear_extrap_feat_down:
    x = df[(df['product_code'] == 10065) & (df['week_infected'] > 0)]['infected_cumsum']
    y = df[(df['product_code'] == 10065) & (df['week_infected'] > 0)][col_name]
    f = interpolate.interp1d(x, y, fill_value='extrapolate')
    xnew = np.array(infected_extrap)
    ynew = f(xnew)
    
    if ynew[1] <= ynew[0]:
        ynew_min = y.min()
        ynew[ynew < ynew_min] = ynew_min
    first_min_extrap = np.where(ynew==y.min())[0]
    if len(first_min_extrap) > 0:
        ind = first_min_extrap[0]
        pattern = np.hstack([y.to_numpy(), ynew[:ind+1]])
        ynew = ynew[:ind]
        while len(ynew) < numweek_extrap:
            infected_on_thisperiod = covid_feat['infected'].to_numpy()[:len(ynew)]
            if (len(infected_on_thisperiod)>1):
                if (infected_on_thisperiod[-1] <= infected_on_thisperiod[-2]):
                    pattern -= bias
                    ynew = np.hstack([ynew, pattern])
                else:
                    ynew = np.hstack([ynew, pattern])
            else:
                ynew = np.hstack([ynew, pattern])
        ynew = ynew[:numweek_extrap]
    test_0[col_name] = ynew
    
#Ð´Ð¾Ð±Ð°Ð²Ð»ÑÐµÐ¼ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ¸ Ð¸Ð· sir Ð¼Ð¾Ð´ÐµÐ»Ð¸
test_0['week_infected'] = covid_feat['infected']
test_0['week_recovered'] = covid_feat['recovered']
test_0['week_deaths'] = covid_feat['deaths']
```

```python
#ÑÐ¾Ð±Ð¸Ñ€Ð°ÐµÐ¼ Ð¿Ð¾Ð»Ð½Ñ‹Ð¹ ÑÐºÑÑ‚Ñ€Ð°Ð¿Ð¾Ð»Ð¸Ñ€Ð¾Ð²Ð°Ð½Ð½Ñ‹Ð¹ Ð´Ð°Ñ‚Ð°ÑÐµÑ‚ ÑÐ¾ Ð²ÑÐµÐ¼Ð¸ Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð°Ð¼Ð¸
all_products = df['product_code'].unique()
test_list = []
for prod in all_products:
    test_1 = test_0.copy()
    test_1['product_code'] = [prod]*numweek_extrap
    cat_code = df[(df['product_code'] == prod)]['category code'].unique()[0]
    test_1['category code'] = [cat_code]*numweek_extrap
    for col_name in product_detail_feature:
        x = df[(df['product_code'] == prod) & (df['week_infected'] > 0)]['infected_cumsum']
        y = df[(df['product_code'] == prod) & (df['week_infected'] > 0)][col_name]
        f = interpolate.interp1d(x, y, fill_value='extrapolate')
        xnew = np.array(infected_extrap)
        ynew = f(xnew)
        
        if ynew[1] <= ynew[0]:
            ynew_min = y.mean()
            ynew[ynew < ynew_min] = ynew_min
        else:
            ynew_max = y.max() * (1+max_growth)
            ynew[ynew > ynew_max] = ynew_max
        
        test_1[col_name] = ynew
    for col_name in notlin_product_detail_feat:
        x = df[(df['product_code'] == prod) & (df['week_infected'] > 0)]['infected_cumsum']
        y = df[(df['product_code'] == prod) & (df['week_infected'] > 0)][col_name]
        f = interpolate.interp1d(x, y, fill_value='extrapolate')
        xnew = np.array(infected_extrap)
        ynew = f(xnew)
    
        if ynew[1] <= ynew[0]:
            ynew_min = y.min()
            ynew[ynew < ynew_min] = ynew_min
            first_min_extrap = np.where(ynew==y.min())[0]
            if len(first_min_extrap) > 0:
                ind = first_min_extrap[0]
                pattern = np.hstack([y.to_numpy(), ynew[:ind+1]])
                ynew = ynew[:ind]
                while len(ynew) < numweek_extrap:
                    infected_on_thisperiod = covid_feat['infected'].to_numpy()[:len(ynew)]
                    if (len(infected_on_thisperiod)>1):
                        if (infected_on_thisperiod[-1] <= infected_on_thisperiod[-2]):
                            pattern -= bias
                            ynew = np.hstack([ynew, pattern])
                        else:
                            ynew = np.hstack([ynew, pattern])
                    else:
                        ynew = np.hstack([ynew, pattern])
                ynew = ynew[:numweek_extrap]
        else:
            ynew_max = y.max() * (1+max_growth)
            ynew[ynew > ynew_max] = ynew_max
        test_1[col_name] = ynew
        
    test_list.append(test_1)
df_test = pd.concat(test_list, ignore_index=True)
```
